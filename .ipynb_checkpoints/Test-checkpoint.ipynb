```python
import os
import cv2
from utils.utils import get_yolo_boxes, makedirs
from utils.bbox import draw_boxes
from tensorflow.keras.models import load_model
import numpy as np
from PIL import Image 
```

```python
from yolo_model import YoloModel
```

```python
my_model = YoloModel()
```

```python
my_model.load_model()
```

```python
image = Image.open('D:/Datasets/har/Prototype_01/ch04_20210413074231_79325.jpg')
```

```python
numpy_image=np.array(image) 
preprocess_image=cv2.cvtColor(numpy_image, cv2.COLOR_RGB2BGR)
preprocess_image2=cv2.cvtColor(numpy_image, cv2.COLOR_RGB2BGR)
```

```python
img, list_pos_boxes, list_non_boxes, num_non_pos, num_pos  = my_model.predict(image)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
for i in range(len(list_pos_boxes)):
    label_str = ''
    label = -1
    
    if list_pos_boxes[i][4] > 0.3:
        if label_str != '': label_str += ', '
        label_str += ('Positive' + ' ' + str(round(list_pos_boxes[i][4]*100, 2)) + '%')
#         print(label_str)


    text_size = cv2.getTextSize(label_str, cv2.FONT_HERSHEY_SIMPLEX, 1.1e-3 * preprocess_image.shape[0], 5)
    # print(text_size)

    width, height = text_size[0][0], text_size[0][1]
#     print(width, height)

    region = np.array([[list_pos_boxes[i][0]-3,        list_pos_boxes[i][1]], 
                       [list_pos_boxes[i][0]-3,        list_pos_boxes[i][1]-height-26], 
                       [list_pos_boxes[i][0]+width+13, list_pos_boxes[i][1]-height-26], 
                       [list_pos_boxes[i][0]+width+13, list_pos_boxes[i][1]]], dtype='int32')  
#     print(region)
    cv2.rectangle(img=preprocess_image, pt1=(list_pos_boxes[i][0], list_pos_boxes[i][1]), pt2=(list_pos_boxes[i][2], list_pos_boxes[i][3]), color=(0, 255, 0), thickness=3)
    cv2.fillPoly(img=preprocess_image, pts=[region], color=[0, 255, 0])
    cv2.putText(img=preprocess_image, 
                text=label_str, 
                org=(list_pos_boxes[i][0]+13, list_pos_boxes[i][1] - 13), 
                fontFace=cv2.FONT_HERSHEY_SIMPLEX, 
                fontScale=1e-3 * preprocess_image.shape[0], 
                color=(0,0,0), 
                thickness=2)
```

```python
for i in range(len(list_non_boxes)):
    label_str = ''
    label = -1
    
    if list_non_boxes[i][4] > 0.3:
        if label_str != '': label_str += ', '
        label_str += ('Non-Positive' + ' ' + str(round(list_non_boxes[i][4]*100, 2)) + '%')
#         print(label_str)


    text_size = cv2.getTextSize(label_str, cv2.FONT_HERSHEY_SIMPLEX, 1.1e-3 * preprocess_image.shape[0], 5)
    # print(text_size)

    width, height = text_size[0][0], text_size[0][1]
#     print(width, height)

    region = np.array([[list_non_boxes[i][0]-3,        list_non_boxes[i][1]], 
                       [list_non_boxes[i][0]-3,        list_non_boxes[i][1]-height-26], 
                       [list_non_boxes[i][0]+width+13, list_non_boxes[i][1]-height-26], 
                       [list_non_boxes[i][0]+width+13, list_non_boxes[i][1]]], dtype='int32')  
#     print(region)
    cv2.rectangle(img=preprocess_image2, pt1=(list_non_boxes[i][0], list_non_boxes[i][1]), pt2=(list_non_boxes[i][2], list_non_boxes[i][3]), color=(0, 0, 255), thickness=3)
    cv2.fillPoly(img=preprocess_image2, pts=[region], color=(0, 0, 255))
    cv2.putText(img=preprocess_image2, 
                text=label_str, 
                org=(list_non_boxes[i][0]+13, list_non_boxes[i][1] - 13), 
                fontFace=cv2.FONT_HERSHEY_SIMPLEX, 
                fontScale=1e-3 * preprocess_image.shape[0], 
                color=(0,0,0), 
                thickness=2)
```
